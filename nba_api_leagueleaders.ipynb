```python
import pandas as pd
import json
import requests
import nba_api as nba
```

**First method**

```python
lst = [x+"-"+y for x,y in list(zip([str(i) for i in range(1984,2018)],[str(x)[1:] if len(str(x)) == 3 else str(x) for x in range(85,119)]))]
df_all = pd.DataFrame()
session = requests.Session()
session.headers.update({'User-Agent': 'Custom user agent'})
for i in lst:
    url = "https://stats.nba.com/stats/leagueleaders?LeagueID=00&PerMode=Totals&Scope=S&Season=" + i + "&SeasonType=Regular+Season&StatCategory=PTS"
    response = session.get(url)
    results = response.json()
    df = pd.DataFrame(results["resultSet"]["rowSet"],columns=results["resultSet"]["headers"])
    df["YEAR"] = i
    df_all = pd.concat([df_all,df])
```

```python
df_all.reset_index(inplace=True)
```

```python
df_all.drop(['index'],axis=1,inplace=True)
```

```python
old_abb_names = dict(zip(['CHH', 'GOS', 'KCK', 'NJN', 'NOH', 'NOK', 'PHL', 'SAN', 'SEA','UTH', 'VAN'],["Charlotte Hornets","Golden State Warriors","Kansas City Kings","New Jersey Nets","New Orleans Hornets","New Orleans/Oklahoma City Hornets","Philadelphia 76ers","San Antonio Spurs","Seattle Supersonics","Utah Jazz","Vancouver Grizzlies"]))

old_franchises = pd.DataFrame.from_dict(old_abb_names, orient='index', columns=['full_name'])
old_franchises = old_franchises.reset_index()
old_franchises.rename(columns={'index':'abbreviation'},inplace=True)
old_franchises['id'] = [str(i) for i in range(0,11)]
old_franchises
```

```python
from nba_api.stats.static import teams
results = teams.get_teams()
active_franchises = pd.DataFrame(results)
franchises = pd.concat([active_franchises,old_franchises],axis=0)
franchises
```

```python
df_all = pd.merge(df_all,franchises, left_on = 'TEAM', right_on = 'abbreviation')
df_all.rename(columns={'full_name':'team_full_name'},inplace=True)
df_all
```

```python
df_all['player_season_id'] = df_all['PLAYER'] + df_all['team_full_name'] + df_all['year']
```
